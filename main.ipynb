```python
import pandas as pd
import numpy as np
from os import getcwd, sep
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing    import MinMaxScaler
from sklearn.feature_selection import SelectKBest, chi2
from pickle                   import dump, load
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.feature_selection import VarianceThreshold
```

1. **Carregue o dataset "creditcard.csv"** e faça uma exploração inicial dos dados. Verifique o número de amostras, o número de características, a distribuição das classes (fraude vs. transações legítimas) e se há algum desbalanceamento.

```python
# Carregar o conjunto de dados
df = pd.read_csv(getcwd()+sep+'creditcard.csv')
df
```

2. **Pré-processamento dos dados:**
- Verifique se existem valores ausentes no dataset e decida como lidar com eles.
- Normalize as características numéricas para garantir que todas tenham a mesma escala.
- Separe os dados em conjuntos de treinamento e teste.

```python
df.isnull().sum()
```

```python
X = df.iloc[:,0:-1]
Y = df.iloc[:,-1]

# Create the normalization scaler
scaler = MinMaxScaler(feature_range=(0, 1))

# Fit the normalization scaler and normalize the train output

Xscaler = scaler.fit_transform(X)

#Ys = scaler.fit_transform(Y)

# Save the normalization scaler
dump(scaler, open(getcwd()+sep+'pca.pkl', 'wb')) 

X_train, X_test, y_train, y_test = train_test_split(Xscaler, Y, test_size=0.2, random_state=42)
```

3. **Engenharia de features:**

```python
# Estabelecer o limiar de variância
limiar_de_variancia = 0.80

# Criar o modelo de seleção de features
selecao = VarianceThreshold(threshold = limiar_de_variancia)

# Selecionar as features
features_selecionadas = selecao.fit_transform(X)

# Verificar a estrutura dos dados selecionados
print("Formato dos dados Originais:", np.shape(X))
print("Formato dos dados após a seleção de features:", np.shape(features_selecionadas))
```

4. **Aplicação da LDA:**
- Realize a LDA para reduzir a dimensionalidade dos dados e projetá-los em um novo espaço de características.
- Avalie a quantidade de informação preservada após a projeção e discuta a importância dessa redução dimensional.

```python
# Aplicar LDA
LDA = LinearDiscriminantAnalysis()
X_train_LDA = LDA.fit_transform(X_train, y_train)
X_test_LDA = LDA.transform(X_test)
print("Informação preservada após LDA:", LDA.explained_variance_ratio_.sum())
```

5. **Aplicação do PCA:**
- Realize o PCA para reduzir a dimensionalidade dos dados e projetá-los em um novo espaço de características.
- Verifique a quantidade de informação preservada após a projeção e discuta a importância dessa redução dimensional.

```python
# Aplicar PCA
pca = PCA()
X_train_PCA = pca.fit_transform(X_train)
X_test_PCA = pca.transform(X_test)
print("Informação preservada após PCA:", pca.explained_variance_ratio_.sum())
```

6. **Treinamento e avaliação dos modelos:**
- Escolha dois algoritmos de aprendizado de máquina adequados para o problema de detecção de fraudes, como SVM (Máquinas de Vetores de Suporte) e Random Forest (Florestas Aleatórias).
- Treine os modelos nos dados originais e compare suas performances usando métricas como acurácia, precisão, recall e F1-score.
- Repita o treinamento e avaliação dos modelos nos dados após a aplicação da LDA e do PCA, e compare novamente as performances.

```python
# Treinar e avaliar os modelos nos dados originais
svm = SVC()
svm.fit(X_train, y_train)
svm_preds = svm.predict(X_test)
print("Desempenho do SVM nos dados originais:")
print("Acurácia:", accuracy_score(y_test, svm_preds))
print("Precisão:", precision_score(y_test, svm_preds))
print("Recall:", recall_score(y_test, svm_preds))
print("F1-score:", f1_score(y_test, svm_preds))

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)
print("Desempenho do Random Forest nos dados originais:")
print("Acurácia:", accuracy_score(y_test, rf_preds))
print("Precisão:", precision_score(y_test, rf_preds))
print("Recall:", recall_score(y_test, rf_preds))
print("F1-score:", f1_score(y_test, rf_preds))
```

```python
# Treinar e avaliar os modelos nos dados após LDA
svm_LDA = SVC()
svm_LDA.fit(X_train_LDA, y_train)
svm_preds_LDA = svm_LDA.predict(X_test_LDA)
print("Desempenho do SVM após LDA:")
print("Acurácia:", accuracy_score(y_test, svm_preds_LDA))
print("Precisão:", precision_score(y_test, svm_preds_LDA))
print("Recall:", recall_score(y_test, svm_preds_LDA))
print("F1-score:", f1_score(y_test, svm_preds_LDA))

rf_LDA = RandomForestClassifier()
rf_LDA.fit(X_train_LDA, y_train)
rf_preds_LDA = rf_LDA.predict(X_test_LDA)
print("Desempenho do Random Forest após LDA:")
print("Acurácia:", accuracy_score(y_test, rf_preds_LDA))
print("Precisão:", precision_score(y_test, rf_preds_LDA))
print("Recall:", recall_score(y_test, rf_preds_LDA))
print("F1-score:", f1_score(y_test, rf_preds_LDA))
```

```python
# Treinar e avaliar os modelos nos dados após PCA
svm_PCA = SVC()
svm_PCA.fit(X_train_PCA, y_train)
svm_preds_PCA = svm_PCA.predict(X_test_PCA)
print("Desempenho do SVM após PCA:")
print("Acurácia:", accuracy_score(y_test, svm_preds_PCA))
print("Precisão:", precision_score(y_test, svm_preds_PCA))
print("Recall:", recall_score(y_test, svm_preds_PCA))
print("F1-score:", f1_score(y_test, svm_preds_PCA))

rf_PCA = RandomForestClassifier()
rf_PCA.fit(X_train_PCA, y_train)
rf_preds_PCA = rf_PCA.predict(X_test_PCA)
print("Desempenho do Random Forest após PCA:")
print("Acurácia:", accuracy_score(y_test, rf_preds_PCA))
print("Precisão:", precision_score(y_test, rf_preds_PCA))
print("Recall:", recall_score(y_test, rf_preds_PCA))
print("F1-score:", f1_score(y_test, rf_preds_PCA))
```

7. **Discussão dos resultados:**
- Analise os resultados obtidos e discuta as melhorias ou piorias na performance dos modelos após a aplicação das técnicas de engenharia de features, LDA e PCA.
- Explique possíveis razões para essas mudanças de desempenho e discuta a utilidade dessas técnicas no contexto da detecção de fraudes.

Com base nos resultados obtidos, pode-se observar as mudanças na performance dos modelos SVM e Random Forest após a aplicação das técnicas de engenharia de features, LDA e PCA.

1. Desempenho após LDA:

* SVM: Houve uma ligeira melhoria na acurácia, precisão e F1-score após a aplicação de LDA. No entanto, o recall permaneceu praticamente o mesmo. Isso sugere que o SVM manteve sua capacidade de detectar corretamente a classe positiva (fraudes) após a aplicação de LDA.
* Random Forest: Após a aplicação de LDA, houve uma redução na precisão, recall e F1-score. Isso indica que o modelo Random Forest teve mais dificuldade em identificar corretamente as transações fraudulentas após a aplicação dessa técnica.

2. Desempenho após PCA:

* SVM: Observa-se uma melhoria na acurácia, precisão e F1-score após a aplicação de PCA. O recall permaneceu praticamente o mesmo. Isso sugere que o SVM foi capaz de encontrar um subespaço de características mais discriminantes após a aplicação de PCA, resultando em uma melhor capacidade de detecção de fraudes.
* Random Forest: Após a aplicação de PCA, o modelo Random Forest apresentou um aumento na precisão e F1-score, enquanto o recall permaneceu quase inalterado. Isso sugere que o Random Forest também se beneficiou da redução de dimensionalidade proporcionada pelo PCA, resultando em uma melhor capacidade de distinguir entre transações legítimas e fraudulentas.

As melhorias ou piorias na performance dos modelos podem ser atribuídas às características das técnicas de engenharia de features utilizadas:

1. LDA (Linear Discriminant Analysis): LDA é uma técnica de redução de dimensionalidade que visa encontrar um subespaço de características que maximize a separação entre classes. No entanto, a aplicação de LDA pode não ser benéfica em todos os casos, especialmente quando a separação entre classes não é bem definida. Isso pode explicar a ligeira melhoria no desempenho do SVM e a redução no desempenho do Random Forest.

2. PCA (Análise de Componentes Principais): PCA é uma técnica de redução de dimensionalidade que visa capturar a maior parte da variabilidade dos dados em um número menor de componentes principais. Ao reduzir a dimensionalidade, o PCA pode eliminar ruídos e redundâncias dos dados, permitindo que os modelos se concentrem nas características mais relevantes. Isso pode explicar as melhorias observadas nos desempenhos do SVM e Random Forest após a aplicação de PCA.

No contexto da detecção de fraudes em transações de cartões de crédito, essas técnicas de engenharia de features podem ser úteis para melhorar a performance dos modelos de Machine Learning. A redução de dimensionalidade pode ajudar a lidar com o problema de alta dimensionalidade dos dados, tornando o treinamento e a inferência mais eficientes. Além disso, ao remover características irrelevantes ou redundantes, as técnicas de engenharia de features podem melhorar a capacidade dos modelos de identificar padrões relevantes associados às transações fraudulentas.

No entanto, é importante ressaltar que a escolha e a aplicação das técnicas de engenharia de features devem ser avaliadas caso a caso. Nem sempre todas as técnicas resultarão em melhorias significativas na performance dos modelos. É essencial realizar experimentações e avaliar os resultados para determinar quais técnicas são mais adequadas para o conjunto de dados específico e o modelo em questão.
